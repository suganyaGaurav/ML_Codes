# Check the validity of Email Address
Given a string, WAP to check if the string is a valid email address or not.

An email is a string separated into two parts by @ symbol, a “personal_info” and a domain_name, that is personal_info@domain_name.


| Character | Description | Example
|----|---|----|
| [ ]  | A set of characters | "a-z"
|  \ | Signals a special sequence. Also used to escape special characters | "\n"
| .  | Any character (except newline character) | "he..o"
| ^  | Starts with | "Dr."
| $  | Ends with | "Singh"
| *  | 0 or more occurance | "bix*"
| +  | 1 or more occurance | "bix+"
| { }  |Exactly the specified number of occurances | "bix{1}"
| \| |Either or | "he\|she"
| \w | Matches any word character | ""
| ( ) |Capture Group

```
# Python program to validate an Email

# import re module - re module provides support for regular expressions
import re

# Make a regular expression
# for validating an Email
regex = '^[a-z0-9]+[\._]?[a-z0-9]+[@]\w+[.]\w{2,3}$'
# for custom mails use: '^[a-z0-9]+[\._]?[a-z0-9]+[@]\w+[.]\w+$'

# Define a function for validating an Email
def check(email):

    # pass the regular expression
    # and the string in search() method
    if(re.search(regex,email)):
        print("Valid Email")

    else:
        print("Invalid Email")



# Enter the email
email = "content@edureka.co"

# calling run function
check(email)

email = "content.team@gmail.com"
check(email)

email = "edureka.co"
check(email)
```
